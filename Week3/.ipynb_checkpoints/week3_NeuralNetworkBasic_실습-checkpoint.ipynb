## Activation Function 
- sigmoid & relu : hidden layer activation function 
- softmax : output layer activation function 

```python
class sigmoid:
    # sigmoid 함수를 작성하세요 
    def forward(x):
        return 1 / (1 + np.exp(-x))
    
    # sigmoid 함수의 미분을 작성하세요
    def backward(x):
        return {1 / (1 + np.exp(-x))}*{np.exp(-x) / (1 + np.exp(-x))}
```

\begin{equation*}
    sigmoid(x) = {1 \over 1+exp(-x)}
\end{equation*}  

\begin{equation*}
    \frac{d}{dx}sigmoid(x) = sigmoid(x)\{1-sigmoid(x)\}
\end{equation*}

```python
class relu:
    # relu 함수를 작성하세요
    def forward(x):
        return np.maximum(0, x)
    
    # relu 함수의 미분을 작성하세요
    def backward(x):
        return np.where(x > 0, 1, 0)
```

\begin{equation*}
    relu(x) = \begin{cases}
            x & \text{if } x > 0\\
    		0 & \text{if } x \le 0\\
    	\end{cases}
\end{equation*}  

\begin{equation*}
    \frac{d}{dx} relu(x) = \begin{cases}
            1 & \text{if } x > 0\\
    		0 & \text{if } x \le 0\\
    	\end{cases}
\end{equation*}

```python
class softmax:
    def forward(z):
        y = []
        for zi in z:
            c = np.max(zi)
            exp_zi = np.exp(zi-c)
            sum_exp_zi = np.sum(exp_zi)
            yi = exp_zi / sum_exp_zi
            y.append(yi)

        return np.array(y)
    
    def backward(p, y) :
        dp = p.copy()
        for dpi, yi in zip(dp, y):
            for k in range(dp.shape[1]):
                if k == yi :
                    dpi[k] -= 1
        return dp
```

## Loss Function

```python
def cross_entropy(p, y):
    loss = []
    for pi, yi in zip(p, y):
        for k in range(p.shape[1]):
            if k == yi:
                loss.append((-1) * (np.log(pi[k] + 1e-8))) # 1e-8을 넣은 이유 : pi[k]값이 0일 경우에 -inf 값이 나올 수 있으므로
    return loss
```

## Layer

```python
class Layer:
    def __init__(self, input_size, output_size, std=1e-4) :
        self.input_size = input_size
        self.output_size = output_size
        self.bias = np.random.randn(output_size)
        self.weight = np.random.randn(input_size, output_size)*std
```

## Neural Network
- 각 메소드와 변수들의 역할을 주석으로 달아주세요! 

```python
class CustomNet:
    # CustomNet을 선언할 때 생성되는 값들입니다.
    def __init__(self, lr=0.0001, epoch=500, batch_size=200):
        self.lr = lr # learning rate를 설정, 앞서 정의한 0.0001을 학습률로 받아 가중치 업데이트에 사용
        self.epoch = epoch # 모델의 반복 횟수인 epoch를 지정해줌.
        self.batch_size = batch_size # 학습할 모델의 크기인 batch size를 지정해줌.
        self.loss_function = cross_entropy # 손실함수로는 앞서 정의한 cross_entropy를 사용.
        self.layers = [] # 층에 대한 정보
        self.activations = [softmax] # 활성화 함수(softmax)
        self.nodes = [] # 노드에 대한 정보
    
    # Layer를 추가할 때 호출합니다
    def addLayer(self, Layer): 
        self.layers.append(Layer) # 입력받은 새로운 Layer를 기존의 layer에 추가
        if not self.nodes: # 만약에 nodes에 값이 없을 경우 즉, 첫 입력의 경우에는
            self.nodes.append(np.zeros(Layer.input_size)) # Layer의 input size 만큼의 zero 배열을 추가해줌.
        self.nodes.append(np.zeros(Layer.output_size))    # 그렇지 않으면 output size 만큼의 zero 배열을 추가해줌.
        
    # Activation Function을 추가할 때 호출합니다
    def addActivation(self, Activation):
        tmp = self.activations.pop() # 활성화 함수 list([softmax])의 마지막 활성함수 추출.
        self.activations.append(Activation) # 새로운 Activation 활성함수 추가.
        self.activations.append(tmp) # 추출한 tmp(softmax)를 마지막에 추가.
        
    # 순전파 함수
    def _forward(self, X):
        self.nodes[0] = X.copy() # zero 배열(input_size에 해당하는)에 입력받은 X로 초기화시켜줌.
        output = X.copy() # 같은 X를 output으로 설정.
        for i in range(len(self.layers)): # layer 수만큼 반복.
            Layer = self.layers[i] # 반복에 따른 i번째 layer를 Layer로 설정.
            Activation = self.activations[i] # 마찬가지로 반복에 따른 i번째 activation을 해당 Activation으로 설정.
            output = np.dot(self.nodes[i], Layer.weight) #node값과 가중치의 행렬 곱으로 output 산출.
            output = output+ Layer.bias # output에 bias(편향)을 더하여 output 수정.
            output = Activation.forward(output) #활성화 함수에 적용한 최종 output 산출.
            self.nodes[i+1] = output # 다음의 node 값으로 output을 초기화시켜줌.
        return output
    
    # 역전파 함수
    def _backward(self, X, output, y) :
        for i in reversed(range(len(self.layers))): # layers의 길이만큼을 반대로 반복문 적용(backward이므로)
            a = self.nodes[i+1] # i+1번째인 다음 node값(다음의 layer값) a 설정
            Layer = self.layers[i] # i번째인 현재의 layer 값 Layer 설정
            Activation = self.activations[i] # i번째 활성화 함수 Activation 설정
            if i+1 == len(self.layers): 
                error = Activation.backward(output, y) # 처음 역전파를 진행할 때의 error 값으로 output과 y를 넣어준 backward 값 설정
            else: 
                error *= Activation.backward(a) # 이후의 layer별로 다음의 node(layer 값)인 a를 넣고 수정한 error 계산
            Layer.weight -= np.dot(error.T, self.nodes[i]).T*self.lr/X.shape[0] # 가중치(weight) 업데이트 진행.
            Layer.bias -= error.sum(axis=0)*self.lr/X.shape[0] # 상수항(bias) 업데이트 진행.
            error = np.dot(error, Layer.weight.T) # error와 weight의 행렬곱으로 error 계산
            
    # Accrucy를 반환합니다
    def _accuracy(self, output, y):
        pre_p = np.argmax(output, axis=1) # output 값중 가장 높은 확률을 갖는 값을 지정해줌.
        return np.sum(pre_p==y)/y.shape[0] # 최종 pre_p과 결과값 y가 같은 즉, 분류가 잘 된 값들의 비율을 의미하는 accuracy를 return.
    
    # 데이터셋에 모델을 fit할때 호출합니다
    def fit(self, X, y, val_X, val_y):
        history = {'val_acc': [],'val_loss': []} # accuaracy(정확도)와 loss값 산출을 위한 사전 생성.
        N = X.shape[0]
        for i in range(self.epoch): # 설정한 epoch 수 만큼 반복문 실행.
            for j in range(N//self.batch_size): # 전체 크기 N을 batch_size만큼 나누어 iteration만큼 반복.
                batch_mask = np.random.choice(N, self.batch_size) # batch_size 만큼의 random 값을 batch_mask로 설정.
                X_batch = X[batch_mask] # batch_mask의 index에 해당하는 X를 X_batch로 설정.
                y_batch = y[batch_mask] # batch_mask의 index에 해당하는 y를 y_batch로 설정
                output = self._forward(X_batch) # X_batch 를 순전파시킨 결과를 output으로 설정.
                self._backward(X_batch, output, y_batch) # 해당 output과 X_batch, y_batch로 역전파 진행.
            
            #accuracy와 loss를 기록해둡시다
            output = self._forward(val_X) 
            history["val_acc"].append(self._accuracy(output, val_y)) # 평가 데이터의 정확도 추가.
            history["val_loss"].append(sum(self.loss_function(output, val_y))) # 평가 데이터의 loss 값 추가.
            
            #중간중간 기록을 찍어볼 때 사용. 적절히 조절해 쓰세요
            if i % 50 == 0:
                print(i, "test accuracy :", history["val_acc"][-1])
                print(i, "test loss     :", history["val_loss"][-1])
        return history
```

```python
# 하이퍼파라미터를 적절히 조절해 뉴럴넷을 선언하세요
nn = CustomNet(lr=0.005, epoch=200, batch_size=400)

# 원하는 만큼 층과 활성화 함수를 쌓아 주세요. 기본적으로 2Layer를 예시로 적어드립니다
nn.addLayer(Layer(784,100))
nn.addActivation(sigmoid)
nn.addLayer(Layer(100,10))
```

## Accuracy, Loss Visualization
- 자유롭게 Accuracy나 Loss를 시각화하여 확인하고 결과를 확인해 보세요! 

1. 기존의 파라미터일 때 learning curve(lr=0.005, epoch=200, batch_size=400)

```python
def learning_curve(history):
    fig, loss_ax = plt.subplots()
    acc_ax = loss_ax.twinx()

    loss_ax.plot(history['val_loss'], 'r', label='val loss')
    loss_ax.set_xlabel('epoch')
    loss_ax.set_ylabel('loss')
    loss_ax.legend(loc='upper left')

    acc_ax.plot(history['val_acc'], 'g', label='val acc')
    acc_ax.set_ylabel('accuracy')
    acc_ax.legend(loc='upper right')

    plt.title('Model Accuracy & Loss')
    plt.show()
    
learning_curve(history)
```

모델의 정확도는 0.8446이며 learning curve에서 loss는 계속 감소하며 accuracy는 계속 증가하므로 훈련이 잘되고 있다고 볼 수 있다.

또한 loss가 덜 감소함에 따라 accuracy 값도 현저히 줄어들었음을 볼 수 있다.
