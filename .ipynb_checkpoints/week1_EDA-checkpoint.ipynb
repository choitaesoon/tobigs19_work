### 2-2-3 수치형 변수 간 변수제거
이후 연속형 변수들의 상관관계를 통해 변수들간의 관계성을 파악한 후 상관관계가 큰 값을 가지면 비슷한 성향을 보인다고 할 수 있으므로 더 중요한 변수만 남겨두고 나머지는 제거해준다.

위의 heatmap 시각화 자료를 통해 각 수치형 변수들간의 상관계수를 살펴보면

1. TotalWorkingYears(경력 기간)와 MonthlyIncome(월 대비 급여 수준)의 상관계수 값 = 0.77
2. YearAtCompany(근속 연수)와 YearsInCurrentRole(현재 역할의 년수)의 상관계수 값 = 0.76
3. YearAtCompany(근속 연수)와 YearWithCurrMananger(현재 관리자와 함께 보낸 시간)의 상관계수 값 = 0.77
4. YearsInCurrentRole(현재 역할의 년수)와 YearWithCurrMananger(현재 관리자와 함께 보낸 시간)의 상관계수 값 = 0.71

로 총 4가지 경우에서 0.7이상의 높은 상관계수를 가짐을 확인할 수 있다.

따라서 이 경우에

1. TotalWorkingYears(경력 기간)와 MonthlyIncome(월 대비 급여 수준)을 두 변수 모두 분석에 포함시키지 않고 TotalWorkingYears(경력 기간)만 분석에 포함시킨다.
2. YearAtCompany(근속 연수)와 YearsInCurrentRole(현재 역할의 년수)와 YearWithCurrMananger(현재 관리자와 함께 보낸 시간)의 세 변수 모두 분석에 포함시키지 않고 세 변수끼리 모두 높은 상관관계를 보이므로 YearAtCompany(근속 연수)만 분석에 포함시킨다.

(실제로 이렇게 상관관계만으로 특정 변수를 포함시키고 제거하는 것은 좋진않다. 하지만 첫번째 경우와 두번째 경우의 두 변수, 세 변수는 나머지 다른 수치형 변수들과 비슷한 상관관계를 보이기 때문에 하나만 남겨두어도 큰 지장이 없을것이라 판단하였다.)

```python
remove_set1 = ['MonthlyIncome','YearsInCurrentRole','YearsWithCurrManager']
num_list = [i for i in num_list if i not in remove_set1]
dat = dat.drop(remove_set1, axis=1)
```

### 2-2-4 범주형 변수 간 변수제거 ( 새로운 파생변수 생성)

범주형 변수들 간의 비슷한 의미를 갖는 변수들에 대해 파생변수를 생성하여 축소시킬 수 있다.

```python
cat_list
```

범주형 변수들의 list를 살펴보면 'EnvironmentSatisfaction(업무 환경에 대한 만족도)','JobSatisfaction(업무 만족도)','RelationshipSatisfaction(대인관계 만족도)'는 모두 만족도를 나타내는 변수이므로 이를 하나로 합친 '총 만족도' 파생변수를 생성시킬 수 있다.

```python
dat['total EnvironmentSatisfaction'] = dat['EnvironmentSatisfaction'] + dat['JobSatisfaction'] + dat['RelationshipSatisfaction']
```

```python
dat.loc[dat['total EnvironmentSatisfaction'] <= 4, 'total EnvironmentSatisfaction'] = 1
dat.loc[(dat['total EnvironmentSatisfaction'] <= 6)&(dat['total EnvironmentSatisfaction'] > 4), 'total EnvironmentSatisfaction'] = 2
dat.loc[(dat['total EnvironmentSatisfaction'] <= 8)&(dat['total EnvironmentSatisfaction'] > 6), 'total EnvironmentSatisfaction'] = 3
dat.loc[(dat['total EnvironmentSatisfaction'] <= 10)&(dat['total EnvironmentSatisfaction'] > 8), 'total EnvironmentSatisfaction'] = 4
dat.loc[dat['total EnvironmentSatisfaction'] >= 10, 'total EnvironmentSatisfaction'] = 5
```

세 만족도 변수들은 1부터 4까지의 값들로 이루어져 있으므로 이를 합치면 3부터 12까지의 값들로 나오게 된다. 이는 총 10개의 변수값으로 이루어져 있으므로 5가지의 만족도 점수로 나타낸 최총 'total EnvironmentSatisfaction' 변수를 생성한다.

### 2-2-5 수치형 변수의 범주형 변수 변환 ( 새로운 파생변수 생성)

```python
dat['Age'].value_counts()
```

수치형 변수 중 나이의 경우에는 세대를 구분지어 나타낼 범주화시킬 수 있다.
흔히 60년 초반까지를 베이비붐 세대, 80년대 까지를 X세대, 90년대 중후반까지를 밀레니엄 세대, 그리고 그 이후인 Z세대로 알려져 있으므로 이를 기준으로 새로운 Generation 변수를 생성해준다.

```python
dat['Generation'] = dat['Age'].apply(lambda x : "Baby Boom" if 59 < x 
                                     else "X generation" if 42 < x < 60 
                                     else "Millennium" if 29 < x < 43 else "Z generation")
```

## 4-1 수치형 범수 시각화

각각의 수치형 변수와 target변수 간의 관계를 시각화를 통해 알아보도록 한다.

### 4-1-1 나이와 퇴사여부 사이의 관계

```python
dat['Attrition'].replace(['Yes','No'], [1,0], inplace=True)
```

```python
table = pd.pivot_table(data = dat, values = 'BusinessTravel',index = 'Age', columns = 'Attrition', aggfunc = len).fillna(0)
table['total'] = table[1] + table[0]
table['Attrition_rate'] = table[1] / table['total'] * 100

plt.figure(figsize=(12,8))
plt.title("나이와 퇴사비율", fontsize=16)
sns.barplot(x = table.index, y = table.Attrition_rate)
plt.show()
```

나이와 퇴사비율 그래프를 확인해보면 나이가 들어가면서 자연스럽게 퇴사비율이 증가하는 것을 확인해볼 수 있다. 또한 특이한 점으로는 상대적으로 30대 전까지의 사람들 퇴사비율이 매우 높게 나타남을 알아볼 수 있다. 이는 곧 회사에 적응하지 못한 사람들이나 이직을 원하는 사람들이 회사를 퇴사하고 다른 곳으로 이직함을 뜻한다.

### 4-1-2 회사와의 거리와 퇴사여부 사이 관계

```python
table = pd.pivot_table(data = dat, values = 'Age', index = 'DistanceFromHome', columns = 'Attrition', aggfunc = len)
table['total'] = table[1] + table[0]
table['Attrition_rate'] = table[1] / table['total'] * 100

plt.figure(figsize=(12,8))
plt.title("회사와의 거리와 퇴사비율", fontsize=16)
sns.barplot(x = table.index, y = table.Attrition_rate)
plt.show()
```

상대적으로 회사와의 거리가 많이 날 수록 퇴사를 할 비율이 증가함을 알아볼 수 있다.

### 4-1-3 시간대비 급여수준과 퇴사여부 사이 관계

```python
table = pd.pivot_table(data = dat,values = 'BusinessTravel', index = 'HourlyRate', columns = 'Attrition', aggfunc = len).fillna(0)
table['total'] = table[1] + table[0]
table['Attrition_rate'] = table[1] / table['total'] * 100

plt.figure(figsize=(12,8))
plt.title("시간대비 급여수준과 퇴사비율", fontsize=16)
sns.barplot(x = table.index, y = table.Attrition_rate)
plt.show()
```

시간에 따른 급여 수준과 퇴사비율의 그래프에서는 급여수준이 높아질수록 퇴사비율이 증가한다고는 볼 수 없지만 비슷한 그래프의 형태가 계속해서 반복함을 볼 수 있다. 향후 분석에 있어서 이러한 추세가 영향을 미칠 수 있다. 

### 4-1-4 연봉 상승률과 퇴사여부 사이 관계

```python
table = pd.pivot_table(data = dat,values = 'BusinessTravel', index = 'PercentSalaryHike', columns = 'Attrition', aggfunc = len).fillna(0)
table['total'] = table[1] + table[0]
table['Attrition_rate'] = table[1] / table['total'] * 100

plt.figure(figsize=(12,8))
plt.title("연봉 상승률과 퇴사비율", fontsize=16)
sns.barplot(x = table.index, y = table.Attrition_rate)
plt.show()
```

일반적으로 연봉 상승률이 낮은 사람들은 그렇지 않은 사람에 비해 더 많이 퇴사하는 것으로 보인다. 하지만 특이하게 연봉 상승률이 높은 쪽에서도 퇴사하는 비율이 많은 구간이 있는데 이는 업무에 대한 스트레스처럼 연봉 상승률을 높이는 대신 과도한 업무로 인해 높은 상승률에도 불구하고 퇴사를 하는 경우일 것이라고 예측해볼 수 있다.

### 4-1-4 연봉 상승률과 퇴사여부 사이 관계

```python
table = pd.pivot_table(data = dat,values = 'BusinessTravel', index = 'TotalWorkingYears', columns = 'Attrition', aggfunc = len).fillna(0)
table['total'] = table[1] + table[0]
table['Attrition_rate'] = table[1] / table['total'] * 100

plt.figure(figsize=(12,8))
plt.title("총 경력과 퇴사비율", fontsize=16)
sns.barplot(x = table.index, y = table.Attrition_rate)
plt.show()
```

위 그래프를 보면 일반적으로 경력이 짧을 때 퇴사를 많이한다는 것을 알 수 있다. 그리고 경력이 많은 사람들이 퇴사를 하게되는 것도 볼 수 있는데 이는 정년퇴임이 아닐까 예측해볼 수 있다.

## 4-2 범주형 변수 시각화

### 4-2-1 성별과 퇴사여부 사이 관계

```python
gen_dat = dat[['Gender','Attrition']].groupby('Gender').sum()
bar = plt.bar(gen_dat.index, gen_dat['Attrition'])
plt.title("성별에 따른 퇴사인원 수")
plt.ylabel("퇴사인원 수")
for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)
plt.show()
```

성별에 따른 퇴사인원 수를 확인해보면 남성이 거의 두배가량 많은 수를 보이고 있다.

### 4-2-2 교육의 정도와 퇴사여부 사이 관계

```python
edu_dat = dat[['Education','Attrition']].groupby('Education').sum()
bar = plt.bar(edu_dat.index, edu_dat['Attrition'],color = ['gold', 'black', 'blue', 'green','red'], alpha=0.5)
plt.title("교육의 정도에 따른 퇴사인원 수")
plt.ylabel("퇴사인원 수")
for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)
plt.show()
```

교육의 정도에 따른 퇴사인원 수는 정규분포의 그래프 형태를 보이고 있다. 높은 수준의 교육을 받은 사람들은 그만큼 퇴사를 잘 하지 않는 반면 가장 인원이 많고 적당한 교육을 받은 사람들에서 높은 퇴사인원을 보이고 있다.

### 4-2-3 업무 분야와 종류에 따른 퇴사여부 사이 관계

```python
dep_dat = dat[['Department','Attrition']].groupby('Department').sum()
bar = plt.bar(dep_dat.index, dep_dat['Attrition'], color = ['blue', 'green','red'], alpha=0.5)
plt.title("업무 분야에 따른 퇴사인원 수")
plt.ylabel("퇴사인원 수")
for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 12)
plt.show()
```

업무분야는 'Human Resources','Research & Develpoment','Sale' 세가지로 나타나고 이중에서 Research & Development에 해당하느 업무를 맡은 사람들의 퇴사인원 수가 가장 높은 것을 확인할 수 있다. 반면 Human Resources의 업무를 맡은 사람들은 거의 대부분 퇴사를 하지 않는다는 것을 볼 수 있다.

### 4-2-4 세대별 퇴사여부 사이 관계

```python
gener_dat = dat[['Generation','Attrition']].groupby('Generation').sum()
gener_series = pd.Series(gener_dat['Attrition'], index=gener_dat.index)
```
