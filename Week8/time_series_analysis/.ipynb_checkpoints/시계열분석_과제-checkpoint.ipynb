```python
samsung_df = yf.download('005930.KS',
                      start='2020-01-01',
                      end='2021-04-21',
                      progress=False)

samsung_df = samsung_df[["Close"]]

samsung_df = samsung_df.reset_index()

samsung_df.columns = ['day', 'price']

samsung_df['day'] = pd.to_datetime(samsung_df['day'])

samsung_df.index = samsung_df['day']
samsung_df.set_index('day', inplace=True)

samsung_df
```

## 1. 모형 식별

먼저 데이터를 시각화해봄으로써 시계열 그래프에 추세, 계절성, 주기가 있는지 파악해본다.

```python
plt.figure(figsize=(18,9))
plt.plot(samsung_df.index, samsung_df["price"], linestyle="-")
plt.xlabel=('Day')
plt.ylabel=('Price')
plt.show()
```

먼저 시각화 자료를 통해서 3가지 정도를 파악할 수 있다.

* 정상성을 만족하지 않는다.
* 주가가 점점 상승하다가 최근에 다시 줄어드는 형태이다.
* 계절성과 주기성은 크게 나타나지 않는다.

그렇다면 정상성 여부를 kpss 검정과 ADF 검정을 통해 통계적으로 확인해보도록 한다.

#### 1. kpss 검정

H0(귀무가설) : 정상성을 만족한다.
H1(대립가설) : 정상성을 만족하지 않는다.

```python
def kpss_test(timeseries, pvalue = .05, regression_option = 'ct'):
    print ('Results of KPSS Test:')
    kpsstest = kpss(timeseries, regression= regression_option)
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
    for key,value in kpsstest[3].items():
        kpss_output['Critical Value (%s)'%key] = value
    print (kpss_output)
    if kpss_output[1] < pvalue:
        print(f"정상시계열이 맞다는 귀무가설을 {pvalue*100}%의 유의수준으로 기각할 수 있으므로 해당 데이터는 정상성을 보장하지 못합니다.")
    else:
        print(f"정상시계열이 맞다는 귀무가설을 {pvalue*100}%의 유의수준으로 기각할 수 없으므로 해당 데이터는 정상성이 보장됩니다.")
        

kpss_test(samsung_df["price"])
```

#### 2. ADF 검정

H0(귀무가설) : 정상성을 만족하지 않는다.
H1(대립가설) : 정상성을 만족한다.

```python
def adf_test(timeseries, pvalue = .05, regression_option = 'ct'):
    print ('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC', regression = regression_option)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
    if dfoutput[1] < pvalue:
        print(f"정상시계열이 아니라는 귀무가설을 {pvalue*100}%의 유의수준으로 기각할 수 있으므로 해당 데이터는 정상성이 보장됩니다.")
    else:
        print(f"정상시계열이 아니라는 귀무가설을 {pvalue*100}%의 유의수준으로 기각할 수 없으므로 해당 데이터는 정상성을 보장하지 못합니다.")

# adf 함수를 통해 출력
adf_test(samsung_df["price"])
```

위의 두 검정 결과를 통해 삼성 주가 시계열 데이터는 정상성을 만족하지 못한다고 파악할 수 있다. 따라서 모형의 정상성을 만족해주기 위해 분산 안정화 변환 혹은 차분을 해줄 필요가 있다. 

#### (추가) AIC를 통한 최적의 ARIMA 모형 찾기

```python
model_arima= auto_arima(samsung_df,trace=True, error_action='ignore', start_p=1,start_q=1,max_p=3,max_q=3,suppress_warnings=True,stepwise=False,seasonal=False)
model_arima.fit(samsung_df)
```

일반적으로 ARIMA 모형의 AIC 통계량(AIC : 통계 모델의 성능을 측정해주는 기준값) 값이 작을수록 더 좋은 모형이라고 판단할 수 있다. 위의 auto_arima를 통해 정상 시계열을 만들기 전에 기존의 samsung_df에서 어떤 ARIMA(p, d, q)모형을 사용해야 할지 미리 예측해보도록 한다.  

ARIMA(0,1,0), ARIMA(1,1,0), ARIMA(0,1,1)의 세 모형에 대해서 가장 낮은 AIC이 나왔다.
우선 세 모형 모두 1차 차분을 보여주고 있기 때문에 비정상 시계열에 대해서 먼저 1차 차분을 진행해보도록 한다.

```python
log_df = np.log(samsung_df.price) # 로그 변환
log_samsung_df = log_df.diff(1).dropna() # 1차 차분
```

```python
log_samsung_df.plot(figsize=(12, 8)) 
plt.show()
```

```python
print("ACF 값 : {0}".format(sm.tsa.stattools.acf(log_samsung_df, nlags=10, fft=False)))
print("PACF 값 : {0}".format(sm.tsa.stattools.pacf(log_samsung_df, nlags=10, method='ywm')))
```

```python
fig = plt.figure(figsize=(8,6))
fig.set_facecolor('white')
ax = fig.add_subplot()
sm.graphics.tsa.plot_acf(log_samsung_df, lags=20, ax=ax)
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
fig.set_facecolor('white')
ax = fig.add_subplot()
sm.graphics.tsa.plot_pacf(log_samsung_df, lags=10, ax=ax)
plt.show()
```

위의 결과로 ARIMA(p, 1, q)모형에 대한 ACF와 PACF 그래프를 확인해보면 AR차수인 p는 PACF 그래프를 통해 1시차에서만 파란색 범위를 벗어나기 때문에 p = 1로, MA차수인 q는 ACF 그래프를 통해 1시차에서만 파란색 범위를 벗어나므로 q = 1-1 = 0으로 설정할 수 있다. 따라서 잠정 모형을 **ARIMA(0,1,0)** 으로 설정할 수 있다.

이떄 과대적합의 가능성으로 차수 p와 q에 각각 1씩 더해준 ARIMA(1,1,0), ARIMA(0,1,1)의 모형에 대해서도 가능성을 염두해두어야 한다. 따라서 총 3개의 모형 중 최적의 모형을 찾기 위해서 앞서 행했던 auto_arima를 통한 AIC값을 비교하여 가장 낮은 값인 모형을 최종 모형으로 선택한다.

```python
model_arima= auto_arima(log_df,trace=True, error_action='ignore',d = 1, start_p=1,start_q=1,max_p=3,max_q=3,suppress_warnings=True,stepwise=False,seasonal=False)
model_arima.fit(log_df)
```

auto_arima를 통해 여러 ARIMA모형들의 AIC값을 비교하여 AIC 값이 가장 낮은 **ARIMA(0,1,0)** 모형을 최종 모형으로 선정한다.
